```python
from ExcitonBSE import ExcitonMoS2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
N=36 # Exciton Hamiltonian mesh
```

```python
gamma=np.array([0.,0.,0.])
M=np.array([0,-0.5,0.])
K=np.array([0.33,-0.33,0.])
kpoints=[];N1=int(N/2);N2=int(N1/2);
for i in range(N1):
    kpoints.append(i*M/N1+(N1-i)*gamma/N1)
for i in range(N2):
    kpoints.append(i*K/N2+(N2-i)*M/N2)
kpoints.append(K)
kpoints=np.array(kpoints);kpoints=np.array(np.round(kpoints*N),dtype=np.int)
```

```python
E=[]
for k in kpoints:
    Exciton=ExcitonMoS2()
    Exciton.constuctExcitonHamiltonian(epsilon=2.5,N=N,vb=[6],cb=[7],shift=0.45,Q=k)
    Exciton.solveExcitonHamiltonian()
    E.append(Exciton.EH)
    print(k,Exciton.EH[:4])
E=np.array(E)
```

```python
plt.figure()
for i in range(100):
    plt.plot(E[:,i],'-+',color='blue')
plt.plot(np.ones(50)*(N1),np.linspace(0,3),'-',color='green')
plt.plot(np.ones(50)*(N1+N2),np.linspace(0,3),'-',color='green')
plt.xticks([0,N1,N1+N2],['$\Gamma$','M','K'])
plt.plot(np.linspace(0,300),np.ones(50)*E[0,0],'--',color='black')
plt.xlim([0.,N1+N2])
plt.ylim([1.85,2.35])
```
