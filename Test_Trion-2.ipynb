```python
from ExcitonBSEKeldysh import ExcitonMoS2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import matplotlib
matplotlib.rc('axes',edgecolor='white')
```

```python
N=[24,30,36]
for n in N:
    Exciton=ExcitonMoS2(epsilon=2.5,N=n,vb=[12,13],cb=[14,15],spin_orbit=True,cutoff=0.6,shift=0.7)
    Exciton.constructTrionBasis()
    Exciton.constructTrionHamiltonian()
    Exciton.constuctExcitonHamiltonian()
    Exciton.calculateAbsobtionSpectrumTrion(eta=0.002)
    Exciton.calculateAbsobtionSpectrum(eta=0.002)
    Exciton.plotAbsobtionSpectrum()
    plt.xlim([2.0,2.4])
    plt.ylim([0,None])
    plt.grid()
    plt.yticks([])
    plt.xlabel('Energy [eV]')
    plt.ylabel('Absorption')
    plt.title('Full:{0}x{0} Cutoff:0.6 With SOC'.format(n))
    plt.savefig('With_SOC/Trion_vs_Exciton_{0}.png'.format(n),dpi=600)
```

```python
N=[24,27,30,36,42]
for n in N:
    Exciton=ExcitonMoS2(epsilon=2.5,N=n,vb=[6],cb=[7],spin_orbit=False,cutoff=0.6,shift=0.7)
    Exciton.constructTrionBasis()
    Exciton.constructTrionHamiltonian()
    Exciton.constuctExcitonHamiltonian()
    Exciton.calculateAbsobtionSpectrumTrion(eta=0.002)
    Exciton.calculateAbsobtionSpectrum(eta=0.002)
    Exciton.plotAbsobtionSpectrum()
    plt.xlim([2.0,2.4])
    plt.ylim([0,None])
    plt.grid()
    plt.yticks([])
    plt.xlabel('Energy [eV]')
    plt.ylabel('Absorption')
    plt.title('Full:{0}x{0} Cutoff:0.6 Without SOC'.format(n))
    plt.savefig('Without_SOC/Trion_vs_Exciton_{0}.png'.format(n),dpi=600)
```
